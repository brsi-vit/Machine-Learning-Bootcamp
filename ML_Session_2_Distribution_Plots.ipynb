UNIVARIATE KDE

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(42)
```

```python
ages = np.random.normal(loc=35, scale=8, size=200)
```

```python
plt.figure(figsize=(7, 4))
```

```python
sns.kdeplot(ages, fill=True, color='steelblue', alpha=0.8)
plt.axvline(np.mean(ages), color='blue', linestyle='--', label=f'Mean = {np.mean(ages):.1f}')
plt.xlabel('Age')
plt.ylabel('Density')
plt.title('Univariate distribution of age')
plt.legend()
plt.show()
```

```python
sns.histplot(ages, kde=True, color='steelblue', bins=20)
plt.title('Histogram + KDE of age')
plt.show()
```

BIVARIATE DISTRIBUTION

```python
np.random.seed(42)
age    = np.random.normal(28, 5, 200)
salary = age * 1200 + np.random.normal(0, 8000, 200)
```

```python
plt.figure(figsize=(7, 5))
sns.kdeplot(x=age, y=salary, fill=True, cmap='Reds', levels=15)
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Bivariate Distribution — Age vs Salary')
plt.show()
```

```python
sns.jointplot(x=age, y=salary, kind='kde', fill=True, cmap='Blues')
plt.suptitle('Joint Distribution of Age and Salary', y=1.01)
plt.show()
```
