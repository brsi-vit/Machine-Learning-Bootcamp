# Box Cox




```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
```

```python
# Simulate data from an exponential distribution (right-skewed)
data = np.random.exponential(scale=2.0, size=1000)
```

```python
# Perform Box-Cox transformation and find the optimal lambda
transformed_data, best_lambda = stats.boxcox(data)
```

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(data, bins=30, color='salmon')
plt.title("Original Skewed Data")
```

```python
plt.subplot(1, 2, 2)
plt.hist(transformed_data, bins=30, color='seagreen')
plt.title(f"Box-Cox Transformed Data (λ = {best_lambda:.2f})")

plt.tight_layout()
plt.show()
```

# Yeo Johnson

```python
import numpy as np
from scipy import stats
from sklearn.preprocessing import PowerTransformer
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
```

```python
data = np.random.exponential(scale=2.0, size=1000).reshape(-1, 1)
```

```python
pt = PowerTransformer(method='yeo-johnson')
data_transformed = pt.fit_transform(data)
print('Optimal lambda:', pt.lambdas_)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
axes[0].hist(data, bins=30, color='salmon')
axes[0].set_title('Original — Right Skewed')
axes[1].hist(data_transformed, bins=30, color='seagreen')
axes[1].set_title(f'Yeo-Johnson Transformed (λ = {pt.lambdas_[0]:.2f})')
plt.tight_layout()
plt.show()
```
